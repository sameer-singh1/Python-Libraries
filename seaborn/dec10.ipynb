```python
df1 = sns.load_dataset('flights')
```

```python
df1
```

```python
x = df1.pivot_table(index='year',columns='month', values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False,linecolor='green',linewidths=1)
```

```python
sns.heatmap(x, cbar=False,linecolor='yellow',linewidths=10,annot=True)
```

```python
sns.heatmap(x, cbar=False,linecolor='yellow',linewidths=10,annot=True,fmt='d')
```

```python
sns.heatmap(x, cbar=False,linecolor='yellow',linewidths=10,annot=True,fmt='d',cmap='spring')
```

```python
sns.heatmap(x, cbar=False,linecolor='yellow',linewidths=10,annot=True,fmt='d',cmap='summer')
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,annot=True,fmt='d',metric='correlation')
```

```python
sns.clustermap(x,annot=True,metric='correlation',z_score=0)
```
